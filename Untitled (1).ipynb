```python
import  numpy as np;
import  pandas as pd;
import  seaborn as sns;
import  matplotlib.pyplot as plt;
```

```python
df = pd.read_csv('/content/Sample_-_Superstore[1].csv',encoding='latin1')
df.head()
```

```python
df.isnull().sum().inplace = 'true'
```

```python
# sales by category
```

```python
df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
```

```python
sns.barplot(x='Category',y='Sales',data=df)
plt.title('Sales by Category')
plt.show()
```

Technology is the highest selling category, followed by Furniture.
Office Supplies has the lowest sales. The company should invest more in Technology products
and improve marketing for Office Supplies to increase revenue.

```python
# profit by segment
```

```python
df.groupby('Segment')['Profit'].sum().sort_values(ascending=False)
```

```python
sns.boxplot(x='Segment',y='Profit',data=df)
plt.title('Profit by Segment')
plt.show()
```

```python
Q1 = df['Profit'].quantile(0.25)
Q3 = df['Profit'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df['Profit'] < lower_bound) | (df['Profit'] > upper_bound)]
print(outliers)
```

The Consumer segment has the highest variability in profit with both the highest profit
and the lowest loss. Corporate and Home Office segments have more stable profits.
Overall, all segments have a median profit around zero, indicating a need to control costs.

```python
# top 10 states by sales
```

```python
sales_state = df.groupby('State')['Sales'].sum().sort_values(ascending=False).head(10)
```

```python
sns.barplot(x=sales_state.index,y=sales_state.values)
plt.title('Top 10 states by sales')
plt.show()
```

California is the highest revenue-generating state with over 450K in sales,
followed by New York and Texas. The top 3 states contribute to more than 50% of total sales.
The company should focus marketing, logistics, and customer service efforts on these key states.

```python
# sales vs profit scatter plot
```

```python
sns.scatterplot(x='Sales',y='Profit',data=df)
plt.title('Sales vs Profit')
plt.show()
```

 Profit generally increases with sales, but there are several orders with high sales
but negative profit. For example, one order with 20K+ sales resulted in a loss of 2000.
This indicates that high discounts or shipping costs are affecting profitability on large orders.

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
monthly_sales = df.groupby(df['Order Date'].dt.to_period('M'))['Sales'].sum()
```

```python
sns.lineplot(x=monthly_sales.index.astype(str),y=monthly_sales.values)
plt.title('Monthly Sales')
plt.xticks(rotation=90)
plt.show()
```

The sales show a consistent upward trend from 2014 to 2017, indicating business growth.
There is strong seasonality with sales peaking in November and December every year.
The company should increase inventory and marketing budget during Q4 to maximize revenue.
